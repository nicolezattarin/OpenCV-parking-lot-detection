```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

saver = tf.train.import_meta_graph('CNN_model/15_epochs_32_batch_classifier/keras_metadata.pb', clear_devices=True)
graph = tf.get_default_graph()
input_graph_def = graph.as_graph_def()
sess = tf.Session()
saver.restore(sess, "CNN_model/15_epochs_32_batch_classifier")
```

```python


a = np.load("CNN_model/15_epochs_32_batch_history.npy", allow_pickle=True).item()

sns.set_theme(style="white", font_scale=2, palette="Dark2")

history_trend = a
epochs = range(1, len(history_trend["accuracy"]) + 1)
fig, ax = plt.subplots(figsize=(10,6))
ax_right=ax.twinx()
lw=3
sns.lineplot(epochs, history_trend['accuracy'], lw=lw, label='train', ax=ax)
sns.lineplot(epochs, history_trend['val_accuracy'], lw=lw, label='validation', ax=ax)
sns.lineplot(epochs, history_trend['loss'], lw=lw, label='train loss', ax=ax_right)
sns.lineplot(epochs, history_trend['val_loss'], lw=lw, label='validation loss', ax=ax_right)
ax.legend(loc='center right')
ax_right.get_legend().remove()
ax.set_xlabel('Epochs')
ax.set_ylabel('Accuracy')
ax_right.set_ylabel('Loss')
ax.set_ylim(0,1)
ax_right.set_ylim(0,1)
fig.savefig("CNN_model/15_epochs_32_batch_history.png", bbox_inches='tight')
```
